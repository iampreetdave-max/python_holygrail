# linked list

linked list is a linear data structure which stores data and memory refrencess in a form of a node
each node contains the refrence of next and or preveious node based upon type of linked list 
the memory locations of stored data are not contiguous in linked list 
linked list can be bifarcated as singly linked list, doubly linked list,circular linked list, doubly circular linked list {extra=headed linked list} 

## singly linked list 
in this link list each node contains the memory refrence to next node 
the refrence of first node is called head pointer and refrence of last node is called null pointer 


### implementation of singly linked list 
obj oriented prograning is utilised to define nodes and following operations can be performed 
1-storing the linked list 
2-traversal of linked list \n
3-insertion of new node 
4-deletion of existing node 


```python
class node:
    def __init__(self,data):
        self.data=data
        self.next=None
n1=node(10)
n2=node(20)
n3=node(30)
n4=node(40)

n1.next=n2
n2.next=n3
n3.next=n4

cnode=n1
print("Head==>",end="")
while cnode:
    print(cnode.data,end="==>")
    cnode=cnode.next
    
print("Null")
```

## circular linked list

in this all nodes are connected in such a way that it creates a close circular path
where last node is connected with first node
it helps in over writing the existing data without changing its index position 

```python
class node:
    def __init__(self):
        self.data=int(input("enter a number:"))
        self.next=None
n1=node()
n2=node()
n3=node()
n4=node()

n1.next=n2
n2.next=n3
n3.next=n4
n4.next=n1

cnode=n2
startnode=n1

print(startnode.data,end="==>")
while cnode!=startnode:
    print(cnode.data,end="==>")
    cnode=cnode.next
    
print(startnode.data)
```

## doubly linked list

```python
class node:
    def __init__(self):
        self.data=int(input("enter element:"))
        self.next=None
        self.prev=None
        
n1=node()
n2=node()
n3=node()
n4=node()

n1.next=n2
n2.next=n3
n3.next=n4
n4.next=n1

n1.prev=n4
n2.prev=n1
n3.prev=n2
n4.prev=n3

snode=n1
cnode=n2
print(snode.data,"<==> ",end="")
while snode!=cnode:
    print(cnode.data,end=" <==>")
    cnode=cnode.next  
```

##### doubly linked list has 3 diffrent versions 
##### 1 modefified doubly linked list with no overwritting of values 
#### 2 doubly linked list with some properties of circular doubly linked list (basic doubly linked list)
##### 3 headed linked list version of doubly linked list(header linked list)

```python
#binary tree
each node in binary tree has max 2 children known as left child and right child 
#binary search tree
it is a type of binary tree in which left child node is less then parent node and 
right child node has greater value then parent node 
#AVL tree
it is a type of binary search tree that self balances every node 
```

```python
class tree:
    def __init__(self,data):
        self.data=data
        self.right=None
        self.left=None
        
r=tree('a')
n1=tree('b')
n2=tree('c')
n3=tree('d')
n4=tree('e')
r.left=n1
r.right=n2
n1.left=n3
n1.right=n4
print("root",r.data)

cnode=r
while cnode.left!=None or cnode.right!=None:
    x=cnode.left
    y=cnode.right
    print("left node",x.data,"right node",y.data)
    cnode=cnode.left
```
